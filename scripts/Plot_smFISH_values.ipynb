```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import math
from scipy import stats
```

```python
# specify the datasets for your plot
df = pd.read_csv('/Users/lbreima/Documents/PhD_local/Manuscripts/2021_Sam_zipcodes/database/total_database.csv') 
```

```python
df.head()
```

```python
map2 = df.loc[df['gene'] == 'Map2']
gapdh = df.loc[df['gene'] == 'GAPDH']
```

```python
sns.reset_defaults()

# Create figure instance
fig, ax = plt.subplots(figsize=(4.5, 5))


#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


# Add boxplot
sns.boxplot(map2['shRNA'],map2['smFISH_per_area'],
            
            hue=map2['structure'], palette=["m", "g"],
            )


sns.stripplot(map2['shRNA'],map2['smFISH_per_area'], hue=map2['structure'], dodge=True)

plt.ylabel('smFISH per area', fontsize=15) 
plt.title('Map2')
#plt.ylim(0.8, 1)
#plt.xlim(-0.1,1.7)
plt.xticks(fontsize=10)
 

#  Show and save the figure
plt.tight_layout()
plt.savefig('smFISH_per_area_map2.png', dpi=300)
plt.show()
```
